```python
def findWord(s):
    return 'collection' in s.lower()
```

```python
df[df['Product Name'].apply(findWord)]['Product Name'].value_counts()
```

```python
a = df[ df['Product Name'].apply(findWord) ]
b = a['Product Name'].value_counts()
lst = b[b == 4]
for x in lst.index:
    display(df[df['Product Name']==x])
```
