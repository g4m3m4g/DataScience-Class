```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dic = {'day': ['Sun','Mon','Tue','Wed','Thu','Fri','Sat'],
      'temp':[24, 25, 29, 35, 23, 20, 23],
      'temp2':[20,14,15,16,47,48,49]}
df = pd.DataFrame(dic)
x = df['day']
y = df['temp']
z = df['temp2']
plt.figure(figsize=[3,2])
plt.plot(x,y)
plt.plot(x,z)
plt.plot(x,y,'s-.r', label='A')
plt.plot(x,y,'o:g', label='B')
plt.legend(loc='best')
plt.show()
```

```python
plt.figure(figsize=[3,2])
plt.plot(x,y,ls=':',lw=2, color=(0.5,0,1), marker='*', ms=10, mfc='green')
plt.title('Temp around this week')
plt.xlabel('Day')
plt.ylabel('Temp')
plt.ylim(20,30)
plt.xticks([0,1,2,3,4,5,6], 'A B C D E F G'.split())
plt.xticks(rotation=70)
plt.yticks(np.arange(17,41,2))
plt.style.use('default')
plt.style.use('Solarize_Light2')
plt.savefig('plotting.png')
plt.show()
```

```python
fig, axes = plt.subplots(ncols=3, nrows=1)
axes[0].plot(x,y,'o-r')
axes[1].plot(x,y,'o-r')
fig.tight_layout()
```

```python
fig = plt.figure()
axes1 = fig.add_axes( [0,0,1,1] )
axes1.plot(x,y)

axes2 = fig.add_axes( [1,1,0.5,0.5] )
axes2.plot(x,z)

axes3 = fig.add_axes( [0.7,0.7,0.3,0.3] )
axes3.plot(x,z)
```
