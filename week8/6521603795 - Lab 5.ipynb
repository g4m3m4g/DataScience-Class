## Exercise 3



**ใช้  arrays x, y และ z เพื่อทำการ plot บนแกนที่สร้างจากข้อที่แล้ว  (Notice อย่าลืมกำหนด  x - limits และ y - limits )**

```python
fig = plt.figure()
axes1 = fig.add_axes( [0,0,1,1] )
axes1.set_xlabel('X')
axes1.set_ylabel('Z')
axes1.plot(x,z)

axes2 = fig.add_axes( [0.2,0.7,0.2,0.2] )
axes2.set_title('Zoom')
axes2.set_ylim(30,50)
axes2.set_yticks(np.arange(30,51,5))
axes2.set_xlim(20,22)
axes2.set_xticks([20.0, 20.5, 21.0, 21.5, 22.0])
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.plot(x,y)
```

## Exercise 4

**จงใช้คำสั่ง plt.subplots(nrows=1, ncols=2)**

**จากนั้นให้ทำการ plot (x,y) และ plot (x,z) บนแกน axes  และให้ใช้งานคำสั่ง linewidth and style เพื่อตกแต่งเส้นของกราฟ**

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=[7, 2.5])
axes[0].plot(x, y, '--b', lw=2)
axes[0].set_yticks(np.arange(0, 201, 50))
axes[1].plot(x, z, 'r', lw=3)
```

#### โค้ดต่อไปนี้ใช้ในสองข้อสุดท้าย

```python
df['Category'].unique()
```

```python
df[ df['Category'] == 'Furniture' ].groupby('Year').sum()['Profit']
```

##### Dictionary of Category

```python
arr_df = {}
for i in range(0,df['Category'].nunique()) :
    arr_df[df['Category'].unique()[i]] = df[ df['Category'] == df['Category'].unique()[i] ].groupby('Year').sum()['Profit']
```

## จงแสดงกราฟรายได้ของแต่ละ Category ในแต่ละปีในกราฟเดียว 

```python
markers = ['^', 'o', 's']
plt.xticks([2014,2015,2016,2017])
for i, (category, profit) in enumerate(arr_df.items()):
    plt.plot(profit.index, profit.values, label=category, marker=markers[i % len(markers)])

plt.legend(title='')
plt.show()
```

## จงแสดงกราฟรายได้ของแต่ละ Category ในแต่ละปี แบบแยกกราฟ

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,0.6,0.6])
axes1.plot(arr_df['Furniture'].index, arr_df['Furniture'], marker='^', label='Funiture',  ls='--', color='blue', lw=3)
axes1.set_title('Furniture')
axes1.set_xticks(np.arange(2014,2018))

axes2 = fig.add_axes([0.5,0.8,0.6,0.6])
axes2.plot(arr_df['Technology'].index, arr_df['Technology'], marker='o', label='Technology',  ls=':', color='g', lw=3)
axes2.set_title('Technology')
axes2.set_xticks(np.arange(2014,2018))

axes3 = fig.add_axes([1,0,0.6,0.6])
axes3.plot(arr_df['Office Supplies'].index, arr_df['Office Supplies'], marker='s', label='Office Supplies',  ls='-.', color='r', lw=3)
axes3.set_title('Office Supplies')
axes3.set_xticks(np.arange(2014,2018))
```
