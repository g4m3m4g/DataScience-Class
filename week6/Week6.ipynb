# I.check if OWN_OCCUPIED has rows that is not 'Y' or 'N'
# II.change value in OWN_OCCUPIED from lower case to upper case
# III.find value that is not 'Y' or 'N' again then change them to 'None'

```python
upper = ['Y','N']

df[ (df['OWN_OCCUPIED'].apply( lambda x : x not in upper))]
```

```python
lower = ['y', 'n']
df[df['OWN_OCCUPIED'].apply(lambda x : x in lower)]
```

```python
def toUpper(word):
    try:
        word.upper()
        return word.upper()
    except:
        return word

df['OWN_OCCUPIED'] = df['OWN_OCCUPIED'].apply(toUpper)
df
```

```python
def changeToNone(x):
    if x in upper:
        return x
    else: 
        return np.nan
        #return None
    
```

```python
df['OWN_OCCUPIED'] = df['OWN_OCCUPIED'].apply(changeToNone)
df
```

```python
df['NUM_BEDROOMS'] = df['NUM_BEDROOMS'].astype('Int64', errors='ignore')
df
```

```python
def fn(x):
    try:
        x = float(x)
        if x > 0:
            return round(x)
        else :
            return 0
    except:
        return np.nan

df['NUM_BATH'].apply(fn).astype('Int64', errors='ignore')
```
