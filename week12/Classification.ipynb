```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('../data/shirt_size.xlsx')
df.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=5)
# x = Attribute
# y = Label
X = df[['Height_cms', 'Weight_kgs']]
y = df['T_Shirt_Size']
```

```python
model.fit(X, y)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = model.predict(X)
```

```python
y_pred
```

```python
acc = accuracy_score(y, y_pred)
acc
```

```python
df2 = pd.DataFrame( {'Label' : y , 'Predict' : y_pred} )
df2
```

```python
df = pd.read_excel('../data/customer.xlsx')
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
```

```python
order = ['Male', 'Female']
OE  = OrdinalEncoder(categories= [order] )
df['gender'] = OE.fit_transform(df[['gender']])
df.head()
```

```python
df2 = pd.DataFrame()
df2 = df.copy()
scaler = StandardScaler()
df2[['gender', 'age', 'salary']] = scaler.fit_transform(df2[['gender', 'age', 'salary']])
df2.head()
```

```python
df2
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df2[['gender', 'age', 'salary']]
y = df2['purchased']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print(acc)
```

```python
pd.DataFrame( {'Label': y_test, 'Predict': y_pred})
```

```python
for i in range(3,20):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    print(f"n = {i}, accuracy = {acc}")
```
