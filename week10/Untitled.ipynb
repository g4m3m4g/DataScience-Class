```python
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(data = df, x = 'sex')
```

```python
sns.histplot(data=df, x = 'tip', bins=20, kde=True, hue='sex')
```

```python
sns.countplot(data = df, x = 'day')
```

```python
sns.countplot(data = df, x = 'sex', hue = 'day', palette='pastel')
```

```python
sns.boxplot(data = df, x = 'day', y = 'tip', hue = 'sex', palette='pastel')
```

```python
sns.violinplot(data = df, x = 'day', y = 'tip', hue = 'sex', palette='pastel')
```

```python
sns.stripplot(data = df, x = 'day', y = 'tip', palette='pastel')
```

```python
sns.stripplot(data = df, x = 'day', y = 'tip', palette='pastel', jitter = 0.2)
```

```python
sns.stripplot(data = df, x = 'day', y = 'tip', palette='pastel')
sns.boxplot(data = df, x = 'day', y = 'tip', hue = 'sex', palette='pastel')
```

```python
sns.swarmplot(data = df, x = 'day', y = 'tip',hue = 'sex')
```

```python
sns.scatterplot(data = df, x = 'total_bill', y = 'tip',hue = 'sex')
```

```python
sns.lineplot(data = df, x = 'total_bill', y = 'tip',hue = 'sex', marker='o')
```

```python
a = df.select_dtypes(include=['float64','int64']).corr()
sns.heatmap(a, cmap='flare', lw=2)
```

```python
g=sns.jointplot(data= df, x='total_bill', y='tip', kind='reg')
g.fig.set_size_inches(3,3)
```

```python
g=sns.JointGrid(data= df, x='total_bill', y='tip')
g.plot(sns.scatterplot, sns.histplot)
g.fig.set_size_inches(3,3)
```

```python
sns.pairplot(df, hue='sex')
```

```python
g = sns.PairGrid(df, hue='sex')
g.fig.set_size_inches(3,3)
g.map.(sns.histplot)
g.map_upper(sns.lineplot)
g.map_lower(sns.lineplot)
```

```python
g = sns.FacetGrid(data = df, col ='day', row='sex')
g.map(sns.scatterplot, 'total_bill', 'tip')
```
