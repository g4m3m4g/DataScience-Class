** ให้นำ data ของ titanic ที่เป็น dataframe ไปใช้ทำการ plot ให้ตรงกับภาพดังต่อไปนี้**

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.pairplot(titanic)
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.jointplot(data= titanic, x='fare', y='age', hue='sex')
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.histplot(data=titanic, x='fare', bins=20, kde=True, color='red')
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.swarmplot(data = titanic, x = 'class', y = 'age', size=4)
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.boxplot(data = titanic, x = 'sex', y = 'age', hue = 'class')
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
plt.figure(figsize=[5,4])
sns.countplot(data = titanic, x='sex', hue='class', palette='Set1')
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
a = titanic.select_dtypes(include=['float64','int64']).corr()
sns.heatmap(a, lw=2, cmap="coolwarm")
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
g = sns.FacetGrid(data = titanic, col ='class', row='sex')
g.map(sns.histplot, 'age', bins=10)
```
